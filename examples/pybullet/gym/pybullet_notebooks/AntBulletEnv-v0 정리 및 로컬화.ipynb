```python
import gym
import numpy as np
from gym.wrappers import TimeLimit
import pybullet
import pybullet_envs
import pybullet_data

from stable_baselines import PPO2
from stable_baselines.common.evaluation import evaluate_policy
```

```python
# Create environment
# 여기를 돌리면 화면이 띄고, 이 화면을 계속 띄워놓아야 학습이 됩니다!
# 안띄워놓으면 후에 화면 안뜸

env = gym.make('AntBulletEnv-v0')
env.render("human")
env.reset()
```

```python
# 이 코드를 활용해 hyperparameter 조정, 지금은 기본값 입니다.
model = PPO2('MlpPolicy', env, learning_rate=1e-3, verbose=1)
```

```python
print(env.spec)
```

```python
# 학습
model.learn(total_timesteps=10000)
```

```python
# 학습된 모델 저장
model.save("ant_bullet")
del model  # delete trained model to demonstrate loading
```

```python
# 여기 주소를 저장 위치로 맞춰주세요
model = PPO2.load("/home/hayley/.mujoco/stable-baseline/ant_bullet")
```

```python
# 학습된 모델로 재생 시작
obs = env.reset()
for i in range(1000):
    action, _states = model.predict(obs)
    obs, rewards, dones, info = env.step(action)
    env.render()
```
